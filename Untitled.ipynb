Solving the transcendental equation in finite square well 

```python
import matplotlib.pyplot as plt 
from scipy.optimize import fsolve
import numpy as np
```

The equation that we have to solve is :
$$\sqrt{(zo/z)^2 -1} = tan(z) $$

```python
zo = 8
```

Plotting $ tan(z) $ function 

```python
z = np.linspace(0, 6 * np.pi, 1000)
y = np.tan(z)

# NaN means "Not a Number" and NaNs are not plotted or connected.  By setting these values
#to NaN, you effectively mark them as invalid or undefined, indicating that they should not be plotted.

y[:-1][np.diff(y) < 0] = np.nan

# show grid
plt.grid()

plt.xlabel("z")
plt.ylabel("$tan(z)$")

# Set the x and y axis cutoffs
plt.ylim(0,10)
plt.xlim(0, 6 * np.pi)

plt.title("$y = tan(x)$", fontsize=14)
plt.plot(z, y)
plt.show()
```

Plotting $\sqrt{(zo/z)^2 -1}$

```python
x = np.sqrt((zo/z)**2 - 1)
plt.grid()
plt.plot(z,x)
plt.show 
```

```python
z = np.linspace(0, 4 * np.pi, 1000)
zo = 15

y = np.tan(z)
x = np.sqrt((zo/z)**2 - 1)
y[:-1][np.diff(y) < 0] = np.nan

fig = plt.figure(figsize=(20, 8)) 

plt.ylim(0,10)
plt.xlim(0, 4 * np.pi)


plt.plot(z,y,'r-', linewidth = 3)
plt.plot(z,x,'b-', linewidth =3);

idx = np.argwhere(np.diff(np.sign(y - x))).flatten()

plt.plot(z[idx], y[idx], 'ks', markersize=20)

plt.xticks(np.linspace(0, 4*np.pi, 9), ['0', 'π/2', 'π', '3π/2', '2π','5π/2' ,'3π','7π/2' , '4π'])
plt.tick_params(axis='both', which='both', labelsize=20)

plt.grid(True)

plt.show()
```
