$$BOUND\  STATE$$
$$Solving\ the\ transcendental\ equation\ for\ finite\ square\ well $$

```python
import matplotlib.pyplot as plt 
from scipy.optimize import fsolve
import numpy as np
```

The equation that we have to solve is :
$$\sqrt{(zo/z)^2 -1} = tan(z) $$

```python
zo = 15         #take any arbitary value of zo
```

Plotting $ tan(z) $ function 

```python
z = np.linspace(0, 4 * np.pi, 1000)    
y = np.tan(z)
y[:-1][np.diff(y) < 0] = np.nan      # This operation inserts a NaN where the difference between
                                     #successive points is negative
                                     # NaN means "Not a Number" and NaNs are not plotted or connected
 


plt.grid()                      #displays grid



plt.xlabel("z")                #labelling the axis 
plt.ylabel("$tan(z)$")



plt.ylim(0,10)                 #setting x and y limits 
plt.xlim(0, 4 * np.pi)



plt.plot(z,y)
plt.show
```

Plotting $\sqrt{(zo/z)^2 -1}$

```python
x = np.sqrt((zo/z)**2 - 1)
plt.grid()
plt.plot(z,x)
plt.show 
```

```python
z = np.linspace(0, 4 * np.pi, 1000)
zo = 15

y = np.tan(z)
x = np.sqrt((zo/z)**2 - 1)
y[:-1][np.diff(y) < 0] = np.nan

fig = plt.figure(figsize=(20, 8))  #setting the ratio of graph to be displayed

plt.ylim(0,10)
plt.xlim(0, 4 * np.pi)

#plotting the the trancendental equation
plt.plot(z,y,'r-', linewidth = 3)
plt.plot(z,x,'b-', linewidth =3);

idx = np.argwhere(np.diff(np.sign(y - x))).flatten()   #finds the intersection points of 2 graphs

plt.plot(z[idx], y[idx], 'ks', markersize=20)   #displaying the intersection points with black squares

plt.xticks(np.linspace(0, 4*np.pi, 9), ['0', '/2', '', '3/2', '2','5/2' ,'3','7/2' , '4']) #displaying x marks
plt.tick_params(axis='both', which='both', labelsize=20)  #

plt.grid(True)

plt.show()
```

Number of bound states = number of intersection points

In this case for zo=15 we have 4 bound states

$$ SCATTERING\  STATE$$
$$ Plotting\ transmission\ coefficient\ as\ a\ function\ of\ energy  $$

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as constant
```

$$T^{-1} = 1 + \frac{{V^2}}{{4E(E+V)}} \sin^2\left(\frac{{2a}}{{\hbar}}\sqrt{{2m(E+V)}}\right)$$

```python
E = np.linspace(0,100,50)   #setting the energy values 

v = 20                      #the value of potential strength depends upon the quantum system 
a = 0.7                     # width of the well
m = constant.electron_mass  # you can take mass of any quantum particle, here is chose the mass of electron
h = constant.hbar           # h is the reduced plancks constant

b = (2*a*(2*m*(E+v))**1/2)/h  # b is the angle of sin function

S = np.sin(b)**2               # S is the sine squared value   

T = 1 + ((v**2)*S)/(4*E*(E+v))  # T is the inverse of transmission coefficient as displayed in equation above
tra_coeff = 1/T                 # tra_coeff is the actual transmission coefficient that we have to plot against energy

fig = plt.figure(figsize=(20, 8)) 

plt.ylim(0,1.1)
plt.xlim(0,100)

plt.grid()
plt.plot(E,tra_coeff,'g-', linewidth = 3)
plt.show 
```
